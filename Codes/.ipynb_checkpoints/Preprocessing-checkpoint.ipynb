```python
Ext = ['_MIPS24_maic.fits', '_CORNISH_5GHz.fits', '_MIPS70_filt.fits']
Folders = ['MIPS24','CORNISH', 'MIPS70', 'GLIMPSE2', 'GLIMPSE3', 'UKIDDS']
```

```python
SorceF = rootdir+'/'+'Fitts'
```

```python
name1 = df.sourceName[1]+ Ext[1]
```

```python
length = len(df)
names = []
Obj_type = []
recta = []
recta_type = []
Edge_Survey = []
Edge_SuType = []
# Cache top-level contents
source_dirs = set(os.listdir(SorceF))

for i in range(length):
    obj = df.sourceType[i]
    name = df.sourceName[i] + Ext[1]

    if obj in source_dirs:
        subdir_path = os.path.join(SorceF, obj)

        try:
            subdir_files = set(os.listdir(subdir_path))
        except FileNotFoundError:
            continue

        if name in subdir_files:
            file_path = os.path.join(subdir_path, name)

            try:
                sh = fits.open(file_path, memmap=True)[0].data

                if (np.isnan(sh).sum() / sh.size) <= (10 / 100):
                    Length, width = np.shape(sh)

                    if Length == width:
                        Obj_type.append(obj)
                        names.append(name)
                    else:
                        recta.append(name)
                        recta_type.append(obj)
                else:
                    Edge_Survey.append(name)
                    Edge_SuType.append(obj)

            except Exception as e:
                print(f"Error processing {file_path}: {e}")
                continue
```
