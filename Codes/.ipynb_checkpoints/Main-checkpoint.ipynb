```python
path = os.getcwd()
rootdir = os.path.dirname(path)
SorceF = os.path.join(rootdir, 'Fitts')
```

```python
df = pd.read_csv(os.path.join(rootdir, 'CN2'))
```

```python
x = list(unique_everseen(df.sourceType))
na = ['PN', 'Radio-Galaxy (Lobe)', 'UCHII', 'Radio-Star', 'Dark HII-Region',
      'Radio-Galaxy (Central Source)', 'Diffuse HII-Region', 'Galaxy', 'IR-Quiet']
```

```python
Ext = ['_MIPS24_maic.fits', '_CORNISH_5GHz.fits', '_MIPS70_filt.fits']
Survey = ['MIPS24','CORNISH', 'MIPS70', 'GLIMPSE2', 'GLIMPSE3', 'UKIDDS']
```

```python
results = sort_fits_files(df, SorceF, Ext[1], na, nan_threshold=0.1)
```

```python
print("Valid square:", len(results["names"]))
print("Rectangular:", len(results["recta"]))
print("Edge cutouts:", len(results["Edge_Survey"]))
```

```python
plt.figure(figsize=(6, 6), dpi=150)
plt.imshow(hd, cmap="gray", origin="lower", interpolation="nearest")
plt.title(f"{results['Edge_SuType'][2]}/{results['Edge_Survey'][2]}")
plt.axis("on")
plt.tight_layout()
plt.show()
```
